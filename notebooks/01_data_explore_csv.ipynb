```python
import numpy
import os
import pandas
import plotly
import pyarrow
import pyarrow.parquet as pq
import sys

plotly.offline.init_notebook_mode(connected=True)
```

```python
# skip: (0-9) excess header rows, (11) units row containing special characters
rows_to_skip = list(range(10))
rows_to_skip.append(11)
print(rows_to_skip)
```

```python
file_path = '../data/2014_Targa_Sixty-Six/20140221_01_02_03_250lm.csv'
```

```python
df = pandas.read_csv(file_path, skiprows=lambda x: x in rows_to_skip)
```

```python
sig_names = ['time', 'handwheelAngle', 'throttle', 'brake', 'latitude', 'longitude', 'altitude', 'horizontalSpeed', 'vxCG', 'vyCG', 'yawAngle', 'pitchAngle', 'rollAngle', 'distance']
```

```python
df[['time', 'handwheelAngle', 'throttle', 'brake', 'latitude', 'longitude', 'altitude', 'horizontalSpeed', 'vxCG', 'vyCG', 'yawAngle', 'pitchAngle', 'rollAngle', 'distance']].head()
```

```python
len(df)
```

```python
data = []

time = df['time'].values

length = 1000

for sig_name in sig_names[1:]:
    trace = plotly.graph_objs.Scatter(
        x = time[:length],
        y = df[sig_name].values[:length],
        name = sig_name
    )
    data.append(trace)

plotly.offline.iplot(data)
```
